### 4.2 Sleep

You would be banned, if you scrape a website too fast. Let your crawler sleep for a while after each round.

```python
import time

for i in range(5):
    time.sleep(3)
    print(i)
```

### 4.3 Randomness

Pausing for extactly three seconds after each round is too robotic. Let's add some randomness to make your crawler looks more like a human.

```python
from random import random

for i in range(5):
    t = 1 + 2 * random()
    time.sleep(t)
    print(i)
```

```python
import requests
import csv
from bs4 import BeautifulSoup

# Define the URL of the Indeed search page for data scientist jobs
url = 'https://www.indeed.com/jobs?q=data+scientist'

# Send a GET request to the URL and retrieve the HTML content
req = requests.get(url)
webpage = req.text

# Create a BeautifulSoup object to parse the HTML
soup = BeautifulSoup(webpage, 'html.parser')

# Find all the job listing elements
job_listings = soup.find_all('div', class_='jobsearch-SerpJobCard')

# Create an empty list to store the job data
data = []

# Iterate over each job listing element
for job in job_listings:
    # Extract relevant data from the job listing
    title = job.find('a', class_='jobtitle').text.strip()
    company = job.find('span', class_='company').text.strip()
    location = job.find('span', class_='location').text.strip()
    summary = job.find('div', class_='summary').text.strip()

    # Append the extracted data as a row to the list
    data.append([title, company, location, summary])

# Save the data as a CSV file
filename = 'data_scientist_jobs.csv'
with open(filename, 'w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerow(['Title', 'Company', 'Location', 'Summary'])  # Write header row
    writer.writerows(data)

print(f"Scraped job data saved as {filename}")





```
